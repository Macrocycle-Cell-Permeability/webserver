```python
from PIL import Image
import os
from rdkit.Chem import Draw
from rdkit import Chem  # Make sure to import Chem
import pandas as pd

# define directory name
dir_name = 'Overall_images'

# check if the directory exists, if not create it
if not os.path.exists(dir_name):
    os.makedirs(dir_name)

# Assuming sample_data is defined earlier in your code
# Drop duplicates
sample_data = pd.read_csv("/Users/qiushifeng/Desktop/macrocycle/macrocycles_overall_mar11.csv")
sample_data.drop_duplicates(subset='Standardise_SMILES', keep="first", inplace=True)

# Create Images and save in target folder
for index, row in sample_data.iterrows():
    molecule = Chem.MolFromSmiles(row['Standardise_SMILES'])
    if molecule is not None:  
        # Set bgcolor=None for transparent background
        img = Draw.MolToImage(molecule, size=(300, 300), bgcolor=None)
        
        # Ensure the image is saved in a format that supports transparency, like PNG
        img.save(os.path.join(dir_name, f"{row['ID']}.png"))
    else:
        print(f"SMILES string at index {index} could not be converted into a molecule.")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

ring_size_counts = Counter(sample_data["Macrocycle_Free_Amide_Ratio"])

sizes = list(ring_size_counts.keys())
counts = list(ring_size_counts.values())

plt.bar(sizes, counts, width=0.4, align='center', color='blue', alpha=0.7)
plt.xlabel('Macrocycle_Overall_Amide_Count')
plt.ylabel('Count')
plt.title('Distribution of Overall Amide Count')
plt.show()
```
